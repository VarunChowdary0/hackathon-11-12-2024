```python
import pandas as pd
import json
```

```python
df = pd.read_csv('./Coding Profile.csv')
```

```python
df.info()
```

# ----------------------- LeetCode Cleaning   ---------------------

```python
new_lc_df = df[['Roll No', 'Leetcode Profile User Id:']]
```

```python
new_json = {}
```

```python
with open('../json/leetcode.json') as file:
    old_json = json.load(file)
```

```python
jsoned_LC = {}
for i in old_json:
    if i['Username']:
        usernm = i['Username'].replace('https://leetcode.com/u','').replace('/','')
        jsoned_LC[i['RollNumber'].upper()] = usernm
    else:
        jsoned_LC[i['RollNumber'].upper()] = ''
```

```python
jsoned_LC
```

```python
for i in range(len(new_lc_df['Roll No'])):
    rollnumber = (new_lc_df['Roll No'][i])
    username = (new_lc_df['Leetcode Profile User Id:'][i])
    jsoned_LC[rollnumber.upper()] = username
```

```python
len(jsoned_LC),len(old_json),f"Additional : {abs(len(jsoned_LC)-len(old_json))} new accounts "
```

```python
with open('../json/leetcode_new.json', "w") as json_file:
    json.dump(jsoned_LC, json_file, indent=4)
```

# ------------------- HackerRank Cleaning ----------------------

```python
df = pd.read_csv('./Coding Profile.csv')
```

```python
df_hackerRank_data = df[['Roll No','Hacker rank profile name']]
```

```python
with open('../json/hackerRank.json') as file:
    old_hackers = json.load(file)
```

```python
all_hacker_formatted = {}

for i in old_hackers:
    all_hacker_formatted[i['RollNumber'].upper()]=i['Username']
```

```python
for i in range(len(df_hackerRank_data['Roll No'])):
    rollnumber = (df_hackerRank_data['Roll No'][i]).upper()
    username = (df_hackerRank_data['Hacker rank profile name'][i]).replace('https://www.hackerrank.com/profile','').replace('/','')
    all_hacker_formatted[rollnumber] = username
```

```python
len(all_hacker_formatted),len(old_hackers),f"Additional : {abs(len(all_hacker_formatted)-len(old_hackers))} new accounts "
```

```python
with open('../json/HackerRank_new.json', "w") as json_file:
    json.dump(all_hacker_formatted, json_file, indent=4)
```
