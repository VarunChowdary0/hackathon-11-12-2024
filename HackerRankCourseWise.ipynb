```python
from bs4 import BeautifulSoup
import requests
import json
```

```python
class hackerRankScrapper:
    def __init__(self) -> None:
        self.headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
                'Referer': 'https://google.com'
            }
    def get_HackerRank(self,username,roll):
        myDct = {   "RollNumber":roll,
                        "badges":{},
                        "certificates":{},
                        "username" : username,
                        "name" : '',
                    }
        if username:
            url = "https://www.hackerrank.com/profile/"+username
            session = requests.Session()
            response = session.get(url,headers=self.headers)
            if response.status_code == 200:
                soup = BeautifulSoup(response.content,'html.parser')
                res = soup.find_all('div',class_='hacker-badge')
                for i in range(len(res)):
                    name = res[i].text
                    re0 = res[i].find('g',class_="star-section")
                    res2 = (re0.find_all('svg',class_="badge-star"))
                    myDct['badges'][name] = len(res2)

                res = soup.find_all('a',class_="certificate-link hacker-certificate")
                for i in range(len(res)):
                    Basic = res[i].find_all("h2")
                    for j in Basic:
                        j = (str(j.text).split(" "))
                        myDct['certificates'][j[1]] = j[2].lstrip('(').rstrip(')')

                        
                name = soup.find('h1',class_="hr-heading-02 profile-title ellipsis")
                myDct['name'] = (name and name.text) or None
                userName = soup.find('p',class_="profile-username-heading hr-body-01 hr-m-t-0.25")
                if userName:
                    myDct['username'] = userName.text.strip("@")
        return myDct
```

```python
with open('json/HackerRank_new.json') as file:
    users = json.load(file)
```

```python
c = 0
n = len(users)
for i in users:
    c+=1
    jsoned.append(hackerRankScrapper().get_HackerRank(users[i],i))
    print(f"{str(c)} / {str(n)}")
```
