```python
from bs4 import BeautifulSoup
import pandas as pd
import json
import requests
```

```python
class hackerRankScrapper:
    def __init__(self) -> None:
        self.headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
                'Referer': 'https://google.com'
            }
    def get_HackerRank(self,username,roll):
        myDct = {   "RollNumber":roll,
                        "badges":{
                            "oneStarBadge": 0,
                            "twoStarBadge": 0,
                            "threeStarBadge": 0,
                            "fourStarBadge": 0,
                            "fiveStarBadge": 0
                        },
                        "certificates":{
                            "basic" : 0,
                            "intermediate":0,
                            "advanced" : 0
                        },
                        "username" : username,
                        "name" : '',
                    }
        if username:
            url = "https://www.hackerrank.com/profile/"+username
            session = requests.Session()
            response = session.get(url,headers=self.headers)
            if response.status_code == 200:
                soup = BeautifulSoup(response.content,'html.parser')
                res = soup.find_all('div',class_='hacker-badge')
                for i in range(len(res)):
                    re0 = res[i].find('g',class_="star-section")
                    res2 = (re0.find_all('svg',class_="badge-star"))
                    if len(res2) == 1:
                        myDct['badges']['oneStarBadge'] += 1
                    if len(res2) == 2:
                        myDct['badges']['twoStarBadge'] += 1
                    if len(res2) == 3:
                        myDct['badges']['threeStarBadge'] += 1
                    if len(res2) == 4:
                        myDct['badges']['fourStarBadge'] += 1
                    if len(res2) == 5:
                        myDct['badges']['fiveStarBadge'] += 1

                res = soup.find_all('a',class_="certificate-link hacker-certificate")
                for i in range(len(res)):
                    Basic = res[i].find_all("h2")
                    for j in Basic:
                        j = (str(j.text).split(" "))
                        if "(Basic)" in j:
                            myDct['certificates']['basic'] += 1
                        elif "(Intermediate)" in j:
                            myDct['certificates']['intermediate'] += 1
                        elif "(Advanced)" in j:
                            myDct['certificates']['advanced'] += 1
                name = soup.find('h1',class_="hr-heading-02 profile-title ellipsis")
                myDct['name'] = (name and name.text) or None
                userName = soup.find('p',class_="profile-username-heading hr-body-01 hr-m-t-0.25")
                if userName:
                    myDct['username'] = userName.text.strip("@")
        return myDct
```

```python
with open('json/HackerRank_new.json') as file:
    users = json.load(file)
```

```python
jsoned_data = []
```

```python
username = []
oneStarBadge =[]
twoStarBadge =[]
threeStarBadge =[]
fourStarBadge =[]
fiveStarBadge =[]
basic_Certificate = []
intermediate_Certificate = []
adavanced_Certificate = []
RollNumber = []
name = []
```

```python
users
```

```python
c = 0
for i in users:
    c+=1
    res = hackerRankScrapper().get_HackerRank(users[i],i)
    jsoned_data.append(res)
    username.append(users[i])
    RollNumber.append(users[i])
    name.append(res['name'])
    oneStarBadge.append(res['badges']['oneStarBadge'])
    twoStarBadge.append(res['badges']['twoStarBadge'])
    threeStarBadge.append(res['badges']['threeStarBadge'])
    fourStarBadge.append(res['badges']['fourStarBadge'])
    fiveStarBadge.append(res['badges']['fiveStarBadge'])
    basic_Certificate.append(res['certificates']['basic'])
    intermediate_Certificate.append(res['certificates']['intermediate'])
    adavanced_Certificate.append(res['certificates']['advanced'])
    print(c)

    
```

```python
df = pd.DataFrame({
    'username':username,
    'RollNumber':RollNumber,
'oneStarBadge':oneStarBadge,
'twoStarBadge':twoStarBadge,
'threeStarBadge':threeStarBadge,
'fourStarBadge':fourStarBadge,
'fiveStarBadge':fiveStarBadge,
'basic_Certificate':basic_Certificate,
'intermediate_Certificate':intermediate_Certificate,
'adavanced_Certificate':adavanced_Certificate
})
```
